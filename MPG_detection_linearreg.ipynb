**Feature engineering**

```python
#converting model_year variable to easier other variable
df["model_year"]=1900+df["model_year"]
from datetime import datetime
todays=datetime.today()
this_year=todays.year
df["age"]=this_year-df.model_year
df.drop(["model_year"],axis=1,inplace=True)
```

```python
df.sample(20)
```

**Modelling**

```python

#define two variables
#one consist of  independent and other consist target variable
x=df.drop(["mpg"],axis=1)
y=df.mpg
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train
```

```python
from sklearn.linear_model import LinearRegression
lrmodel=LinearRegression()
lrmodel.fit(x_train,y_train)
lrmodel.intercept_
coef=pd.DataFrame(lrmodel.coef_,index=x.columns,columns=["coefficient"])
coef
```

```python
y_pred=lrmodel.predict(x_test)
mydict={"actual":y_test,"pred":y_pred}
compare=pd.DataFrame(mydict)
compare.sample(10)
```

**Model Evaluation**

```python
from sklearn.metrics import mean_squared_error as mse,mean_absolute_error as mae,r2_score as r2
def eval(actual,pred):
  MAE=mae(actual,pred)
  MSE=mse(actual,pred)
  RMSE=np.sqrt(MSE)
  R2=r2(actual,pred)
  return print(R2,MAE,MSE,RMSE)
```

```python
from yellowbrick.regressor import PredictionError,ResidualsPlot
visualizer=PredictionError(lrmodel)
visualizer.fit(x_train,y_train)
visualizer.score(x_test,y_test)
visualizer.show()
```

```python
visualizer=ResidualsPlot(lrmodel)
visualizer.fit(x_train,y_train)
visualizer.score(x_test,y_test)
visualizer.show()
```
